# Part A

Gold price prediction has become a prominent application of artificial intelligence (AI) and machine learning methodologies. Much like the use of machine learning in email and spam filtering, gold price prediction involves building models on datasets of historical gold prices, incorporating various features such as market indicators, economic data, and other relevant factors. Linear Regression (LR) is a commonly employed approach, wherein the model learns the relationship between these features and the gold prices, enabling it to make predictions on future price movements. Additionally, Bayesian methods have been explored in the context of gold prediction, assessing the likelihood of certain market conditions based on historical data. Hybrid approaches, combining different techniques like LR, Bayesian methods, and Support Vector Machines (SVMs), have been investigated to enhance prediction accuracy. In recent years, deep learning methods, including Convolutional Neural Networks (CNNs) and Recurrent Neural Networks (RNNs), have gained traction for gold price prediction. CNNs can be utilized to analyze visual features, such as charts and patterns, while RNNs consider the sequential nature of time-series data, capturing temporal dependencies for more accurate predictions. Although AI-based models have shown efficacy in forecasting gold prices, ongoing research aims to improve their accuracy and robustness, adapting to evolving market dynamics and ensuring reliable predictions in the face of changing economic conditions.

# Part B

**1. Random Forest Regression**

Random Forest Regression is a powerful ensemble learning technique widely used in predicting gold prices. It excels in handling complex relationships within the data by constructing multiple decision trees and combining their predictions. The model's robustness and resistance to overfitting make it suitable for capturing intricate patterns in the gold market. However, managing a large number of trees may increase the risk of overfitting, and careful tuning of hyperparameters is essential for optimal performance.

**2. Support Vector Machines (SVMs)**

Support Vector Machines have proven effective in gold price prediction, particularly when dealing with non-linear relationships in the data. SVMs work well in scenarios with a clear margin of separation between different classes, making them suitable for capturing complex trends in the gold market. However, SVMs may face challenges with large datasets, and the performance is highly sensitive to the choice of hyperparameters. Careful tuning and consideration of computational resources are crucial for achieving optimal results.

**3. Long Short-Term Memory (LSTM) Networks:**

LSTM networks, a type of recurrent neural network (RNN), are well-suited for time-series forecasting, making them valuable in predicting gold prices. Their ability to capture long-term dependencies in sequential data allows them to discern subtle trends and patterns in the market. However, training deep neural networks like LSTMs can be computationally intensive and may require a substantial amount of data to generalize well. Despite these challenges, LSTMs offer a powerful tool for gold price prediction, especially when dealing with dynamic and evolving market conditions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

# Data collection and Pre-Processing

```python
gold_data = pd.read_csv("/content/gld_price_data.csv")
```

```python
gold_data.head(10)
```

```python
# spliting features (other coloumn than gold and date) and target
X = gold_data.drop(['Date','GLD'],axis = 1) # whenver droping column, for rows mention zero
Y = gold_data['GLD']
```

```python
print(X)
```

```python
print(Y)
```

```python
# spliting training and testing
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
#trainig
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(x_train, y_train)
```

# Part D

# Evaluation of the trained model

```python
# evalution
test_data_pred =  regressor.predict(x_test)
```

```python
print(test_data_pred)
```

```python

error_score = metrics.r2_score(y_test, test_data_pred)
print("Error values:", error_score)
```

**Prediction vs Actual Output**

```python
import matplotlib.pyplot as plt
import numpy as np



y_test = np.array(y_test)

bar_width = 0.4

plt.bar(np.arange(len(y_test)) - bar_width/2, y_test, width=bar_width, color='green', label='Actual Values', alpha=0.7)
plt.bar(np.arange(len(test_data_pred)) + bar_width/2, test_data_pred, width=bar_width, color='red', label='Predicted Values', alpha=0.7)

plt.title("Actual Price vs Predicted")
plt.xlabel("Num of values")
plt.ylabel("Gold Price")
plt.legend()
plt.show()
```

The testing focused on the Random Forest Regression model for gold price prediction demonstrated a high degree of accuracy, as evidenced by the close alignment between expected and predicted values at the code level. While specific testing on Linear Regression (LR) was not carried out in this instance, the literature review suggests that Random Forest Regression generally outperforms simpler models like LR in capturing complex relationships within the data. The implemented Random Forest Regression model, validated through empirical testing, reinforces its reliability and effectiveness for forecasting gold prices, aligning with the broader insights obtained from the literature review.
